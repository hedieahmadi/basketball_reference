```python
import pandas as pd
```

```python
query1 = """
SELECT players.player_name, players.height
FROM players
JOIN player_season_stats ON players.player_id = player_season_stats.player_id
JOIN seasons ON player_season_stats.season_id = seasons.season_id
JOIN awards ON players.player_id = awards.player_id
WHERE awards.award_name = 'Michael Jordan Trophy'
AND seasons.start_year BETWEEN 2019 AND 2024
AND seasons.end_year BETWEEN 2019 AND 2024
ORDER BY player_season_stats.pts DESC
LIMIT 50
"""
df1 = pd.read_sql(query1, basketball3_etesal)
df1.to_csv('1.csv', index=False)
```

```python
query2 = """
SELECT players.player_name, players.height, players.experience
FROM players
JOIN player_season_stats ON players.player_id = player_season_stats.player_id
JOIN seasons ON player_season_stats.season_id = seasons.season_id
WHERE seasons.start_year IN (2023, 2024)
AND seasons.end_year IN (2023, 2024)
ORDER BY players.experience DESC
LIMIT 15
"""
df2 = pd.read_sql(query2, basketball3_etesal)
df2.to_csv('2.csv', index=False)
```

```python
query3 = """
SELECT players.player_name, player_season_stats.pts, seasons.start_year
FROM players
JOIN player_season_stats ON players.player_id = player_season_stats.player_id
JOIN seasons ON player_season_stats.season_id = seasons.season_id
WHERE seasons.start_year IN (2023, 2022, 2021, 2020)
AND players.player_name IN (SELECT player_name FROM awards WHERE award_name = 'Michael Jordan Trophy')
ORDER BY player_season_stats.pts DESC
LIMIT 10
"""
df = pd.read_sql(query3, basketball3_etesal)
df.to_csv('3.csv', index=False)
```

```python
query4 = """
SELECT players.player_name, players.height, players.weight, seasons.start_year
FROM players
JOIN player_season_stats ON players.player_id = player_season_stats.player_id
JOIN seasons ON player_season_stats.season_id = seasons.season_id
WHERE seasons.start_year IN (2023, 2022)
ORDER BY players.height / players.weight DESC
LIMIT 20
"""
df = pd.read_sql(query4, basketball3_etesal)
df.to_csv('4.csv', index=False)
```

```python
query5 = """
SELECT players.player_name, players.age, player_season_stats.experience, seasons.start_year
FROM players
JOIN player_season_stats ON players.player_id = player_season_stats.player_id
JOIN seasons ON player_season_stats.season_id = seasons.season_id
ORDER BY player_season_stats.experience DESC
LIMIT 10
"""
df = pd.read_sql(query5, basketball3_etesal)
df.to_csv('5.csv', index=False)
```
