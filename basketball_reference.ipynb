```python
import requests
import re
import time
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WIN64; x64)'}
```

```python
url1= "https://www.basketball-reference.com/teams/"
response = requests.get(url1, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# اسم باشگاه
tables = [
    soup.find("table", id="teams_active"),
    soup.find("table", id="teams_defunct")
]
active_teams = tables[0].find_all("a")
defunct_teams = tables[1].find_all("a")
both_teams = active_teams + defunct_teams

def team_names(link):
    names = []
    for n in link:
        names.append(n.text)
    return names

print(team_names(active_teams))
print(team_names(defunct_teams))
print(team_names(both_teams))
print("---------------------------")
print(len(team_names(active_teams)))
print(len(team_names(defunct_teams)))
print(len(team_names(both_teams)))
```

```python
# همه اسم های جدید و قدیمی

active_and_defunct_teams = soup.find_all("th", {"data-stat":"team_name"})

def team_names(link):
    names = []
    for n in link:
        names.append(n.text)
    return names

print(team_names(active_and_defunct_teams))
```

```python
# همه بازی ها
tables = [
    soup.find("table", id="teams_active"),
    soup.find("table", id="teams_defunct")
]


teams_and_games = []

for table in tables:
    rows = table.find_all("tr")
    for row in rows:
        teams = row.find("a")
        games = row.find("td", {"data-stat": "g"})
        if teams and games :
            teams_and_games.append(f"{games.text}")

for i in teams_and_games:
    print(i)
```

```python
#Wins

tables = [
    soup.find("table", id="teams_active"),
    soup.find("table", id="teams_defunct")
]
wins_list = []

for table in tables:
    rows = table.find_all("tr")
    for row in rows:
        teams = row.find("a")
        wins = row.find("td", {"data-stat": "wins"})
        if teams and wins :
            wins_list.append(f"{wins.text}")

for i in wins_list:
    print(i)
```

```python
# Losses

tables = [
    soup.find("table", id="teams_active"),
    soup.find("table", id="teams_defunct")
]
losses_list = []

for table in tables:
    rows = table.find_all("tr")
    for row in rows:
        teams = row.find("a")
        losses = row.find("td", {"data-stat": "losses"})
        if teams and losses :
            losses_list.append(f"{losses.text}")

for i in losses_list:
    print(i)
```

```python
# Win-Loss percentage

tables = [
    soup.find("table", id="teams_active"),
    soup.find("table", id="teams_defunct")
]
w_l_p_list = []

for table in tables:
    rows = table.find_all("tr")
    for row in rows:
        teams = row.find("a")
        win_loss_per = row.find("td", {"data-stat": "win_loss_pct"})
        if teams and win_loss_per :
            w_l_p_list.append(f"{win_loss_per.text}")

for i in w_l_p_list :
    print(i)
```

```python
# playoffs
tables = [
    soup.find("table", id="teams_active"),
    soup.find("table", id="teams_defunct")
]
playoffs_list = []

for table in tables:
    rows = table.find_all("tr")
    for row in rows:
        teams = row.find("a")
        playsoffs = row.find("td", {"data-stat": "years_playoffs"})
        if teams and playsoffs :
            playoffs_list.append(f"{playsoffs.text}")

for i in playoffs_list :
    print(i)
```

```python
# conference championship
tables = [
    soup.find("table", id="teams_active"),
    soup.find("table", id="teams_defunct")
]
conf_list = []

for table in tables:
    rows = table.find_all("tr")
    for row in rows:
        teams = row.find("a")
        conf = row.find("td", {"data-stat": "years_conference_champion"})
        if teams and conf :
            conf_list.append(f"{conf.text}")

for i in conf_list :
    print(i)
```

```python
#League championship

tables = [
    soup.find("table", id="teams_active"),
    soup.find("table", id="teams_defunct")
]
champ_list = []

for table in tables:
    rows = table.find_all("tr")
    for row in rows:
        teams = row.find("a")
        champ = row.find("td", {"data-stat": "years_league_champion"})
        if teams and champ :
            champ_list.append(f"{champ.text}")

for i in champ_list :
    print(i)
```

```python
# تاریخ تاسیس باشگاه
# year_min (From)
# Leagues
table = soup.find("table")
find = table.find("tbody").find_all("tr")

teams = []
leagues = []
founded = []
max = []

for i in find:
    team_name = i.select_one('a[href^="/teams/"]')
    league_name = i.find('td',{'data-stat':'lg_id'})
    founded_year = i.find('td',{'data-stat':'year_min'})
    max_year = i.find('td',{'data-stat':'year_min'})
    
    if team_name and league_name and founded_year and max_year:
        teams.append(team_name.get_text(strip=True))
        leagues.append(league_name.get_text(strip=True))
        founded.append(founded_year.get_text(strip=True))
        max.append(max_year.get_text(strip=True))

print(teams)
print(leagues)
print(founded)
print(max)

```

```python
url_coaches = "https://www.basketball-reference.com/coaches/NBA_stats.html"
response = requests.get(url_coaches, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# اسم مربی ها
table = soup.find("table", {"id":"coaches"})
href = table.find_all("a")
def coach_name(tables):
    coaches = []
    for c in tables:
        coaches.append(c.text)
    return coaches

print(coach_name(href))
```

```python
url_awards = "https://www.basketball-reference.com/awards/"
response = requests.get(url_awards, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
# افتخارات
award_div = soup.find("div", id="content")
awards_link = award_div.find_all("a")

def all_awards(awards_link):
    awards_list = []
    for a in awards_link:
        if a['href'].startswith("/awards/"):
            awards_list.append(a.text)

    return awards_list

print(all_awards(awards_link))
```

```python
# Location
url_players_teams = "https://www.basketball-reference.com"
response = requests.get(url_players_teams + "/teams/")
soup = BeautifulSoup(response.text, "html.parser")

table = [
    soup.find("table", id="teams_active"),
    soup.find("table", id="teams_defunct")
]

active_teams = table[0].find_all("a")
defunct_teams = table[1].find_all("a")
both_teams = active_teams + defunct_teams

def team_url(links):
    https = []
    for l in links:
        href = l.get("href")
        url = url_players_teams + href + "players.html"
        https.append(url)
    return https

active_player_link = team_url(active_teams)
defunct_player_link = team_url(defunct_teams)
all_players_link = active_player_link + defunct_player_link
print(all_players_link)



def get_location(links):
    location_list = []
    for l in links:
        href = l.get("href")
        teams_link = url_players_teams + href
        response = requests.get(teams_link)
        soup = BeautifulSoup(response.text, "html.parser")

        find = soup.find_all("strong")
        for i in find:
            if i.text == "Location:":
                location = i.next_sibling.strip()
                location_list.append(f"{location}")
    return location_list


for item in get_location(both_teams):
    print(item)
```
