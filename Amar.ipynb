```python
import numpy as np  
import pandas as pd 
from scipy import stats 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Players_info.csv')

df.info()
```

```python
df.head(20)
```

Farzie aval

```python
df['agility']=df['Height']/df['Weight']
```

```python
#players= df[((df['From']<=2024)&(df['To']>=2020))|(df['Status']=='Active')].copy()
```

```python
A= df[(df['From']>=2022)&(df['To']<=2024)].copy()
B= df[(df['From']>=2020)&(df['To']<=2022)].copy()
print(f"Amean= {A['agility'].mean():.3f}")
print(f"Bmean= {B['agility'].mean():.3f}")

#print(len(A))
#print(len(B))

stat_A,p_A=stats.shapiro(A['agility'])
stat_B,p_B=stats.shapiro(B['agility'])
print(f"Shapiro A p-value: {p_A:.3f}")
print(f"Shapiro B p-value: {p_B:.3f}")

stat_var,p_var=stats.levene(A['agility'],B['agility'])
print(f"Levene p-value: {p_var:.3f}")

print("t-test: ")
t_stat,p= stats.ttest_ind(A['agility'],B['agility'],alternative='greater')
print(f"t={t_stat:.3f}")
print(f"p={p:.3f}")
if p<0.05:
    print("Afzayesh chaboki!")
else:
    print("tafavote manadari dar chaboki do grouh nist!")
```

```python
fig,axes =plt.subplots(1,2,figsize=(10,4))

axes[0].boxplot([B['agility'],A['agility']],labels=['B','A'])
axes[0].set_ylabel('Agility')
axes[0].set_title('Box Plot')

axes[1].hist(B['agility'],alpha=0.5,label='B',edgecolor='black',density=True,color="#6066d6")
axes[1].hist(A['agility'],alpha=0.5,label='A',density=True,edgecolor='black',color="#60cad6")
axes[1].set_xlabel('Agility')
axes[1].set_ylabel('Density')
axes[1].legend()
axes[1].set_title('Distribution')

plt.tight_layout()
plt.show()
```

Farziye dovom

```python
df['innate']=df['Experience']/df['Age']
A2 =df[(df['From']>=2022)&(df['To']<=2024)].copy()
B2 =df[(df['From']>=2020)&(df['To']<=2022)].copy()

print(f"Amean= {A2['innate'].mean():.3f}")
print(f"Bmean= {B2['innate'].mean():.3f}")
stat_A2,p_A2=stats.shapiro(A2['innate'])
stat_B2, p_B2= stats.shapiro(B2['innate'])
print(f"Shapiro A p-value: {p_A2:.3f}")
print(f"Shapiro B p-value: {p_B2:.3f}")

stat_var2,p_var2= stats.levene(A2['innate'],B2['innate'])
print(f"Levene p-value: {p_var2:.3f}")

print("t-test: ")
t_stat2,p2=stats.ttest_ind(A2['innate'],B2['innate'],alternative='greater')
print(f"t={t_stat2:.3f}")
print(f"p={p2:.3f}")

if p2<0.05:
    print("Behbood tavanaii zati!")
else:
    print("Adame behbood tavanaii zati!")
```

```python
fig, axes = plt.subplots(1,2,figsize=(10,4))
axes[0].boxplot([B2['innate'], A2['innate']], labels=['B','A'])
axes[0].set_ylabel('Tavanaii Zati')
axes[0].set_title('Box Plot ')

axes[1].hist(B2['innate'],alpha=0.5,label='B',density=True,edgecolor='black',color='#FF6B6B')
axes[1].hist(A2['innate'],alpha=0.5,label='A',density=True,edgecolor='black',color='#4ECDC4')
axes[1].set_xlabel('Tavanaii Zati')
axes[1].set_ylabel('density')
axes[1].legend()
axes[1].set_title('Toziie Tavanaii Zati')

plt.tight_layout()
plt.show()
```
